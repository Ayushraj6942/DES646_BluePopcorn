```python
# Simple Python Test: Printing a Pyramid
# No libraries, no data, just text.

def print_pyramid(rows):
    print(f"Building a pyramid with {rows} rows:\n")
    for i in range(rows):
        # Print spaces
        print(' ' * (rows - i - 1), end='')
        # Print stars
        print('*' * (2 * i + 1))

# Run the function
print("--- START OF TEST ---")
print_pyramid(10)
print("--- END OF TEST ---")
```
